# Visulaization

```python
#Box Plot of Budget Distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(y='Budget', data=df)
plt.title('Budget Distribution')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.show()
```

```python
#Area Plot of Collection Over Time
plt.figure(figsize=(10, 6))
df['Collection'].plot(kind='area')
plt.title('Collection Over Time')
plt.xlabel('Time')
plt.ylabel('Collection')
plt.show()
```

```python
#Histogram of Movie Length
plt.figure(figsize=(8, 6))
plt.hist(df['Movie_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Movie Length')
plt.xlabel('Movie Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#Pie Chart of Genre Distribution
plt.figure(figsize=(8, 8))
genre_counts = df['Genre'].value_counts()
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.axis('equal')
plt.show()
```

```python
#Bar Chart of Average Collection by Genre
import matplotlib.pyplot as plt

avg_collection_by_genre = df.groupby('Genre')['Collection'].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_collection_by_genre.plot(kind='bar', color='skyblue')
plt.title('Average Collection by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Collection')
plt.xticks(rotation=45)
plt.show()
```

```python
#Line Chart of Budget Over Tim
plt.figure(figsize=(10, 6))
df['Budget'].plot(kind='line', marker='o', color='green')
plt.title('Budget Over Time')
plt.xlabel('Time')
plt.ylabel('Budget')
plt.grid(True)
plt.show()
```

```python
#Scatter Plot of Budget vs. Collection
plt.figure(figsize=(8, 6))
plt.scatter(df['Budget'], df['Collection'], color='red', alpha=0.5)
plt.title('Budget vs. Collection')
plt.xlabel('Budget')
plt.ylabel('Collection')
plt.show()
```
